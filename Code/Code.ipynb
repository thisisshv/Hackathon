```python
import numpy as np
import pandas as pd
```

### Accounts

```python
accounts["City"] = accounts["City"].fillna(accounts["City"].mode()[0])
```

```python
accounts["ZiPCODE"] = accounts["ZiPCODE"].fillna(accounts["ZiPCODE"].mean())
```

```python
accounts = accounts.drop(["BranchID", "beneficiary", "Area", "State", "Region", "ZiPCODE"],axis=1)
```

```python
accounts.isna().sum()
```

### Transaction

```python
transaction = pd.read_csv("../datasets/transaction.csv")
transaction
```

```python
transaction = transaction.drop(["SELF-INDICATOR", "MATCH-TYPE", "OWNERSHIP-IND", "REPORTED DATE - HIST", "DPD - HIST", "CUR BAL - HIST", "AMT OVERDUE - HIST", "CONTRIBUTOR-TYPE", "DATE-REPORTED", "DISBURSED-DT", "INSTALLMENT-FREQUENCY", "INSTALLMENT-AMT", "AMT PAID - HIST", "ASSET_CLASS", "CLOSE-DT", "LAST-PAYMENT-DATE", "CREDIT-LIMIT/SANC AMT", "OVERDUE-AMT", "TENURE", "WRITE-OFF-AMT"], axis=1)
```

```python
transaction["CURRENT-BAL"] = transaction["CURRENT-BAL"].fillna(transaction["CURRENT-BAL"].mode()[0])
```

```python
transaction = transaction.drop_duplicates()
```

```python
transaction['DISBURSED-AMT/HIGH CREDIT'] = pd.to_numeric(transaction['DISBURSED-AMT/HIGH CREDIT'].str.replace(',', '', regex=False))
transaction['CURRENT-BAL'] = pd.to_numeric(transaction['CURRENT-BAL'].str.replace(',', '', regex=False))
```

```python
transaction
```

```python
def mode_func(x):
    return x.mode().iloc[0]
```

```python
transaction = transaction.groupby('ID').agg(
    ACCT_TYPE=('ACCT-TYPE', mode_func),
    ACCOUNT_STATUS=('ACCOUNT-STATUS', mode_func),
    DISBURSED_AMT_or_HIGH_CREDIT=('DISBURSED-AMT/HIGH CREDIT', lambda x: round(x.mean(), 2)),
    CURRENT_BAL=('CURRENT-BAL', lambda x: round(x.mean(), 2))
)
```

### Spliting x and y

```python
df.columns
```

```python
x = df.drop(["Feedback", "ACCOUNT_STATUS", "DISBURSED_AMT_or_HIGH_CREDIT"], axis=1)
y = df["Feedback"]
```

### Split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7)
```

### Training model

```python
from sklearn.linear_model import LogisticRegression

def create_model_logistic_regression():
    model = LogisticRegression(max_iter=1000)
    model.fit(x_train, y_train)
    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_model_knn():
    # create the model
    model = KNeighborsClassifier(n_neighbors=5)
    
    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
from sklearn.svm import SVC

def create_model_svm():
    model = SVC()
    model.fit(x_train, y_train)
    return model
```

```python
from sklearn.tree import DecisionTreeClassifier

def create_model_decision_tree():
    model = DecisionTreeClassifier(max_depth=50)
    model.fit(x_train, y_train)
    return model    
```

```python
from sklearn.ensemble import RandomForestClassifier

def create_model_random_forest():
    model = RandomForestClassifier(n_estimators=100)
    model.fit(x_train, y_train)
    return model
```

```python
from xgboost import XGBClassifier

def create_model_xgboost():
    model = XGBClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
from sklearn.ensemble import AdaBoostClassifier

def create_model_ada_boost():
    model = AdaBoostClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
from lightgbm import LGBMClassifier

def create_model_lightgbm():
    model = LGBMClassifier()
    model.fit(x_train, y_train)
    return model
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model):
    # training accuracry
    y_pred = model.predict(x_train)
    y_true = y_train

    # calculate training accuracy
    accuracy = accuracy_score(y_true, y_pred)

    # testing accuracy
    y_pred = model.predict(x_test)
    y_true = y_test

    # get the performance metrics for testing
    testing_precision = precision_score(y_true, y_pred,average='micro')
    testing_recall = recall_score(y_true, y_pred,average='micro')
    testing_f1 = f1_score(y_true, y_pred,average='micro')

    return accuracy, testing_precision, testing_recall, testing_f1
```

```python
models = [
    ("Logistic Regression", create_model_logistic_regression()),
    ("KNN", create_model_knn()),
    ("SVM", create_model_svm()),
    ("Decision Tree", create_model_decision_tree()),
    ("Random Forest", create_model_random_forest()),
    ("LGBM", create_model_lightgbm()),
    ("Ada Boost", create_model_ada_boost()),
    ("XGBoost", create_model_xgboost())
]
```

```python
performance_data = []
for model_name, model in models:
    accuracy, precision, recall, f1 = evaluate_model(model)
    performance_data.append([
        model_name, f"{accuracy * 100:0.2f}%", precision, recall, f1
    ])
    
performance_chart = pd.DataFrame(performance_data, 
            columns=["Model", "Accuracy", "Precision", "Recall", "F1"])
performance_chart
```
