```python
import pandas as pd
```

### Accounts

```python
accounts = pd.read_csv("../datasets/accounts.csv")
accounts
```

```python
accounts.isna().sum()
```

```python
accounts["Area"] = accounts["Area"].fillna(accounts["Area"].mode()[0])
```

```python
accounts = accounts.dropna(subset="MaturityDAte")
```

```python
accounts["AGE"] = accounts["AGE"].fillna(accounts["AGE"].mean())
```

```python
accounts["MonthlyIncome"] = accounts["MonthlyIncome"].fillna(accounts["MonthlyIncome"].mean())
```

```python
accounts["City"] = accounts["City"].fillna(accounts["City"].mode()[0])
```

```python
accounts["ZiPCODE"] = accounts["ZiPCODE"].fillna(accounts["ZiPCODE"].mean())
```

```python
accounts = accounts.drop(["BranchID", "beneficiary", "Area", "State", "Region", "ZiPCODE"],axis=1)
```

```python
accounts.isna().sum()
```

### Transaction

```python
transaction = pd.read_csv("../datasets/transaction.csv")
transaction
```

```python
transaction = transaction.drop(["REPORTED DATE - HIST", "DPD - HIST", "CUR BAL - HIST", "AMT OVERDUE - HIST", "CONTRIBUTOR-TYPE", "DATE-REPORTED", "DISBURSED-DT", "INSTALLMENT-FREQUENCY", "INSTALLMENT-AMT", "AMT PAID - HIST", "ASSET_CLASS", "CLOSE-DT", "LAST-PAYMENT-DATE", "CREDIT-LIMIT/SANC AMT", "OVERDUE-AMT", "TENURE", "WRITE-OFF-AMT"], axis=1)
```

```python
transaction["CURRENT-BAL"] = transaction["CURRENT-BAL"].fillna(transaction["CURRENT-BAL"].mode()[0])
```

```python
transaction = transaction.drop_duplicates()
```

```python
transaction
```

```python
def mode_func(x):
    return x.mode().iloc[0]
```

```python
transaction = transaction.groupby('ID').agg(
    SELF_INDICATOR=('SELF-INDICATOR', mode_func),
    MATCH_TYPE=('MATCH-TYPE', mode_func),
    ACCT_TYPE=('ACCT-TYPE', mode_func),
    OWNERSHIP_IND=('OWNERSHIP-IND', mode_func),
    ACCOUNT_STATUS=('ACCOUNT-STATUS', mode_func),
    DISBURSED_AMT_or_HIGH_CREDIT=('DISBURSED-AMT/HIGH CREDIT', mode_func),
    CURRENT_BAL=('CURRENT-BAL', mode_func)
)
```

### Merge

```python
df = pd.merge(accounts, transaction, on=["ID"])
df
```
